### PlateCarree Projection

L-OTI = Land-Ocean Temperature Index

```python
def temp_anomaly_plot(projection, projection_name):
    
    fig = plt.figure(figsize = (12, 6))
    ax = plt.axes(projection = projection)
    ax.set_global()
    ax.gridlines(linestyle = "--", color = "black")

    #Set contour levels, then draw the plot and a colorbar
    clevs = np.arange(-4, 5)
    cmap = "coolwarm"

    #Plot filled contours
    plt.contourf(lons, lats, 
                 temp_anomaly,
                 clevs,
                 transform = ccrs.PlateCarree(),
                 cmap = cmap             
                )
    plt.title(f"L-OTI (C) Anomaly in 2022 vs 1951-1980 \n Projection: {projection_name}")
    
    cb = plt.colorbar(ax = ax,
                      orientation = "horizontal",
                      pad = 0.02,
                      aspect = 20,  #ratio of long to short dimension
                      shrink = 0.5 #Fraction by which to multiply the size of the colorbar
                     )

    cb.set_label("C", size = 12, rotation = 0, labelpad = 15)
    cb.ax.tick_params(labelsize = 10)
    ax.coastlines(resolution = "110m", lw = 1)
    plt.savefig(f"../output/temp_anomaly_{projection_name}.jpeg",
               dpi = 300)

    plt.show()
    
```

```python
ccrs.PlateCarree()
```

```python
temp_anomaly_plot(projection = ccrs.PlateCarree(), projection_name = "PlateCarree")
```

### Robinson Projection

```python
ccrs.Robinson()
```

```python
temp_anomaly_plot(projection = ccrs.Robinson(), projection_name = "Robinson")
```

### Orthographic Projection

```python
ccrs.Orthographic(central_longitude = 90,
                 central_latitude = 90)
```

```python
temp_anomaly_plot(projection = ccrs.Orthographic(central_longitude = 90,central_latitude = 90),
                 projection_name = "Orthographic")
```

### Zonal means data

```python
file = "../data/zonal_means.txt"

import pandas as pd

df = pd.read_csv(file,
                skiprows = 3,
                #sep = " "
                )

#df[['A', 'B']] = df['AB'].str.split(' ', 1, expand=True)

df[["Latitude", "Zonal Mean"]] = df["lat  skip"].str.split("    ", 1, expand = True)


df = df.iloc[:, 1:]

df = df.astype("float")

df.set_index("Latitude", inplace = True)

df
```

```python
df.dtypes
```

```python
labels = np.arange(-90, 91, 15)
labels
```

```python
len(labels)
```

```python
plt.rcParams["font.size"] = 16

df.plot(figsize = (18, 8),
        lw = 0.8,
       color = "red",
       marker = "o",
       markerfacecolor = "black",
       markeredgecolor = "black")

plt.xticks(ticks = np.arange(-90, 91, 15),
          labels = np.arange(-90, 91, 15)
          )

plt.ylabel("Zonal Mean (C)")

plt.legend([], frameon = False)

plt.ylim(0, 3.0)
plt.xlim(-90, 90)

plt.savefig("../output/zonal means data.jpeg",
           dpi = 300)
plt.show()
```
