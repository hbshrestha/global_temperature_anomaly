Calculating 95% confidence interval:
    
    
ci = $\bar x$ $\pm$ z * $\sigma$/ $\sqrt{n}$

The critical z-score values when using a 95 percent confidence level are -1.96 and +1.96 standard deviations.

A z-score (also called a standard score) tells how far the data point is from the mean on a normal distribution. https://www.statisticshowto.com/probability-and-statistics/z-score/

A standard normal distribution is a special normal distribution with mean 0 and standard deviation 1. In a standard normal distribution, 68% of data lie within the 1 standard deviation, 95% of data lie within the 2 standard deviations, and 99.7% of data lie within the 3 standard deviations from the mean. 





## Global surface temperature anomaly plotted as line plots

```python
fig, ax = plt.subplots()
df["Annual Mean"].plot(ax = ax, c = "black", marker = "s")
df["Lowess Smoothing"].plot(ax = ax, c = "red")

x = df.index.tolist()
y = df["Annual Mean"].tolist()

# Define the 95% confidence interval
ci = np.mean(y) + 1.96 * np.std(y) / np.sqrt(len(y))
plt.fill_between(x, y-ci, y+ci,
             color = "gray",
                alpha = 0.25,
                label = "LSAT+SST Uncertainty")

ax.axhline(y = 0, linestyle = "--", color = "black")
plt.legend(loc = "upper left")
plt.ylabel("Temperature anomaly w.r.t. 1951-80 mean (°C)")
plt.title("Global surface air temperature change estimates based on land and ocean data")
plt.savefig("../output/global annual mean surface air temperature change.jpeg",
           dpi = 300)

plt.show()
```

### Test for stationary of time-series
https://machinelearningmastery.com/time-series-data-stationary-python/


Null hypothesis (H0): Time series is not stationary. It has some time dependent feature.

Alternate hypothesis (H1): Time series is stationary. It does not have time-dependent feature.

<b>p-value</b>: 
p-value is the probability of obtaining results at least as extreme as the observed results of a statistical hypotheis test, assuming that the null hypothesis is correct. 

A higher p-value implies that there is stronger evidence in favor of the null hypothesis.
On the other hand, a lower p-value implies that there is stronger evidence in favor of the aleternate hypothesis. This means, that the results are more important/significant.

```python
def test_stationarity(timeseries):
    from statsmodels.tsa.stattools import adfuller

    #ADF test
    result = adfuller(timeseries)

    adf_statistic = result[0]
    p_value = result[1]
    print ("ADF Statistic: ", adf_statistic)
    print ("p-value: ", p_value)

    if p_value > 0.05:
        print ("We fail to reject the null hypothesis. Time series is not stationary, i.e. it has time-dependent features.")

    else:
        print ("Reject the null hypothesis. Time series is stationary, i.e. it does not have time-dependent features.")
```

```python
test_stationarity(df["Annual Mean"])
```
