```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import cartopy.crs as ccrs
from cartopy.feature import NaturalEarthFeature
import xarray as xr
import rioxarray
import numpy as np

file = "../data/amaps_robinson_1200km.nc"
#file = "gistemp1200_GHCNv4_ERSSTv5.nc"
xr_df = xr.open_dataset(file)
xr_df
```

```python
anomaly = xr_df["TEMPANOMALY"]
anomaly
```

```python
cbar_kwargs = {
    "orientation":"horizontal",
    "fraction": 0.045,
    "pad":0.01,
    "extend":"neither"
}

fig = plt.figure(figsize = (20, 20))

ax = fig.add_subplot(1, 1, 1, projection = ccrs.PlateCarree())

ax.add_feature(NaturalEarthFeature("cultural", "admin_0_countries", "10m",
                                  facecolor = "none",
                                  edgecolor = "black"))

ax.set_extent([-150, 150, -55, 85])

anomaly.plot.imshow(ax = ax,
                   add_labels = False,
                   add_colorbar = True,
                   vmin = -4,
                   vmax = 4,
                   cmap = "coolwarm",
                   cbar_kwargs = cbar_kwargs)

ax.set_title("Temperature Anomaly in 2021 [°C] vs 1951-1980")

plt.show()
```
